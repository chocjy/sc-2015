```python
import numpy as np
import h5py
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.axes_grid1 import ImageGrid

msi_filename = '/Users/oruebel/Devel/openmsi-data/msidata/Podophyllum_hexandrum_CS.h5'
cx_data_file = 'spatial_cords.txt'
```

# Get data to generate images from file

```python
msi_file = h5py.File(msi_filename, 'r')
msi_data = msi_file['/entry_0/data_0/data_1']
mz1 = slice(20190, 20210)
mz2 = slice(47300, 47320)
mz3 = slice(28097, 28117)
msimage_1 = msi_data[:,:,mz1].max(axis=2)
msimage_2 = msi_data[:,:,mz2].max(axis=2)
msimage_3 = msi_data[:,:,mz3].max(axis=2)
```

# Load the CX results

```python
cx_data = np.loadtxt(open(cx_data_file,"rb"),delimiter=', ')
x_col = 0
y_col = 1
```

```python
maxx = cx_data[:,x_col].max()
maxy = cx_data[:,y_col].max()
imgdat = np.empty(shape=(maxx,maxy), dtype='float')
imgdat[:] = np.nan
```

# Map of Leverage Score Values

```python
norm = mpl.colors.Normalize()
plt.figure(figsize=(10,10))
imgdat[cx_data[:,x_col].astype(int)-1, cx_data[:,y_col].astype(int)-1] = cx_data[:,3]
plt.imshow(imgdat.transpose(), norm=norm)
plt.colorbar()
plt.title('Map of leverage scores')
```

```python
norm = mpl.colors.Normalize()
plt.figure(figsize=(10,10))
imgdat[cx_data[:,x_col].astype(int)-1, cx_data[:,y_col].astype(int)-1] = np.power(cx_data[:,3], 0.2)
plt.imshow(imgdat.transpose(), norm=norm)
plt.colorbar()
plt.title('Map of leverage scores normalized with power 0.2')
```

# Map of the rank of leverage scores

```python
norm = mpl.colors.Normalize()
plt.figure(figsize=(10,10))
imgdat[cx_data[:,x_col].astype(int)-1, cx_data[:,y_col].astype(int)-1] = np.arange(cx_data.shape[0], dtype='int')
plt.imshow(imgdat.transpose(), norm=norm, cmap='PuBuGn')
plt.colorbar()
plt.title('Map of the rank of leverage scores')
```

# Plot the leverage scores

```python
top_num = [100, 1000, 10000, cx_data.shape[0]]
mean_leverage_score = np.mean(cx_data[:,3])
for i in top_num:
    plt.figure(figsize=(7,4))
    line1 = plt.plot(cx_data[0:i,3], linewidth=2, c='red', label='Leverage Scores')
    plt.ylim(0,cx_data[1,3])
    plt.title('Top %i Leverage Scores' % i)
    line2 = plt.axhline(y = mean_leverage_score, linewidth = 2 , c='blue', label='Mean Leverage Score')
    plt.legend([line1[0], line2])
```

# Top N Locations

```python
top_num = range(100,1001,100) + range(2000,10001, 1000)
fig = plt.figure(1, (18., 18.))
grid = ImageGrid(fig, 111, # similar to subplot(111)
                nrows_ncols = (4, 5), # creates 2x2 grid of axes
                axes_pad=0.25, # pad between axes in inch.
                )
for index, num in enumerate(top_num):
    #plt.figure(figsize=(10, 10))
    grid[index].imshow(np.dstack((msimage_1, msimage_2, msimage_3)))
    #plt.imshow(msimage_2)
    grid[index].scatter(cx_data[0:num,x_col], cx_data[0:num,y_col], c='red' , s=10)
    grid[index].set_title("Top "+str(num)+" locations")
#grid.title("Locations of the top N leverage scores")
```

# Randomized Selecion of Locations

```python
num_pixels = 100   # Number of pixels to be selected

# Compute the accumlative sum of the leverage scores and normalize
bins = np.add.accumulate(cx_data[:,3])
# Normalize the bins by dividing by the max value
bins /= np.max(bins)

# Plot the accumulative leverage score probabilities
plt.figure(figsize=(7,5))
plt.plot(bins)
plt.title('Sampling Weights')
plt.show()

# Generate the plotting matrix
fig = plt.figure(1, (18., 18.))
grid = ImageGrid(fig, 111, # similar to subplot(111)
                nrows_ncols = (3, 3), # creates 2x2 grid of axes
                axes_pad=0.25, # pad between axes in inch.
                )

# Plot static selection of the top nPixels
grid[0].imshow(np.dstack((msimage_1, msimage_2, msimage_3)))
grid[0].scatter(cx_data[0:num_pixels, x_col], cx_data[0:num_pixels, y_col], c='red' , s=20)
grid[0].set_title("Top "+str(num_pixels)+" locations")

# Compute random selections
for i in range(8):
    # Select n random samples based on the distribution of leverage scores 
    infPixelIndices = np.digitize(np.random.random_sample(num_pixels), bins)
    c = cx_data[infPixelIndices,x_col]
    r = cx_data[infPixelIndices,y_col]

    #plt.figure(figsize=(7, 7))
    grid[i+1].imshow(np.dstack((msimage_1, msimage_2, msimage_3)))
    grid[i+1].scatter(c, r, c='r', s=20)
    grid[i+1].set_title("Random sampling, trial {0}".format(i+1))
    #plt.show()
```
